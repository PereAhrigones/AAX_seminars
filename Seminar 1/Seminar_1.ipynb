```python
#Load the data
import pandas as pd
data = pd.read_csv('dataset_Seminar1.txt', sep=' ', header=None)
display(data.head())
```

```python
#Añadimos labels a la data porque me vuelvo loco. (Perdón por comentarios en varios idiomas voy a ir cambiando según el último idioma que haya leído)
data.columns = ["Num of STAs", "Throughput", "Size(x)", "Size(y)", "Area(xy)", "Contention Window", 
                "Channel Width", "Packet Size", "Max RSSI", "Avg RSSI", "Min RSSI", "Avg Prob of Failure", 
                "Throughput", "Average Delay", ]
```

```python
# 3. Mostrar estadísticas básicas (Requisito Parte I) [cite: 11]
print("Estadísticas del dataset:")
display(data.describe().loc[['mean', 'std']])

display(data.head())
```

```python
# 1. Calcular Q1 (25%) y Q3 (75%) para cada columna numérica
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# 2. Definir límites (el factor 1.5 es el estándar de la industria)
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# 3. Filtrar el dataset: solo nos quedamos con filas dentro de los límites
data_clean = data[~((data < limite_inferior) | (data > limite_superior)).any(axis=1)]

print(f"Registros originales: {len(data)}")
print(f"Registros tras limpieza: {len(data_clean)}")
print(f"Outliers eliminados: {len(data) - len(data_clean)}")
```
