```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Si no carga esto en memoria, nada de lo de abajo funciona.
skel_data = pd.read_csv('AAX_seminars\Seminar_2\dataset_Seminar2\Test_skelletonpoints.csv', header=None)
labels_data = pd.read_csv('Train_labels.csv', header=None)
```

```python
# 1. DEFINICIÓN DE LA FUNCIÓN (Adaptada a su CSV de 54 columnas)
def plot_pose(skeleton_row, label, connections=None):
    """
    Grafica la postura humana basada en un array de formato COCO (18 puntos).
    
    :param skeleton_row: Fila del dataframe (array 1D). Primeros 18 valores son X, siguientes 18 son Y.
    :param label: Etiqueta numérica o string de la postura.
    :param connections: Lista de tuplas con los índices de los puntos a conectar.
    """
    plt.figure(figsize=(6, 8))
    
    # Extraer X e Y de forma contundente sabiendo cómo está estructurado su archivo
    x_coords = skeleton_row[0:18]
    y_coords = skeleton_row[18:36]
    
    plt.gca().invert_yaxis()
    plt.scatter(x_coords, y_coords, c='red', s=40, zorder=3, label='Joints')
    
    if connections:
        for inicio, fin in connections:
            plt.plot([x_coords[inicio], x_coords[fin]], 
                     [y_coords[inicio], y_coords[fin]], c='blue', linewidth=2, zorder=2)
                     
    plt.title(f"Pose (Label: {label})", fontsize=16, fontweight='bold')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axis('equal')
    plt.show()

# ==========================================
# 2. INVOCACIÓN DE LA FUNCIÓN CON SUS DATOS
# ==========================================

# Conexiones estándar del esqueleto humano de 18 puntos
conexiones_estandar = [(0,1), (1,2), (2,3), (3,4), (1,5), (5,6), (6,7), 
                       (1,8), (8,9), (9,10), (1,11), (11,12), (12,13), 
                       (0,14), (0,15), (14,16), (15,17)]

# Seleccionamos la instancia que queremos visualizar (ejemplo: la fila 0)
instancia_idx = 0

# Extraemos la fila de los puntos y la etiqueta correspondiente
# OJO: labels_data.iloc[0, instancia_idx] porque sus labels están en una sola fila (transpuestos)
mis_puntos = skel_data.iloc[instancia_idx].values
mi_etiqueta = labels_data.iloc[0, instancia_idx]

# Ejecutamos la función
plot_pose(mis_puntos, mi_etiqueta, conexiones_estandar)
```
